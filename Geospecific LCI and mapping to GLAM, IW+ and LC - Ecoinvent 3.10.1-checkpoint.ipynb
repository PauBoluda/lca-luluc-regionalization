# Exporting regionalized LCI

Next, we'll look at how to extract the regionalized inventory on a single case (the example above).

```python
@lru_cache(maxsize=16000) # lru_cache speeds up repeated lookups. Do not omit it or you might run out of memory and crash.
def cached_lookup(key):
    return bd.get_activity(key)
```

Define a folder where you will store your inventories:

```python
prod_path=r"path where you want to save the file"
```

First, we define a simple function to look up activities in ecoinvent in an efficient way.

To export the regionalized inventory, we need to check each possible combination of elementary flow and activity in our LCA. If the amount of a specific elementary flow for a specific activity is not zero, then we look up information for this activity in the ecoinvent database, such as name, category, unit, reference product, and, in particular: location.

```python
def regio_inventory(lca):
    inventory_results = {'amount':[], 'elementary_flow':[], 'EF_compartment':[], 'EF_unit':[], 'activity_name':[], 'activity_location':[], 'reference_product':[]}
    maxrow=len(lca.biosphere_dict)-1

    for row_key, row in lca.biosphere_dict.items(): # iterate over elementary flows
        clear_output(wait=True)
        print('reading elementary flow number ',row, 'of ', maxrow)
        for col_key, col in lca.activity_dict.items(): # iterate over activities
            if lca.inventory[row, col] != 0: # check in the full inventory matrix if the value for this elementary flow and activity is not zero
                ef=cached_lookup(row_key) # look up the elementary flow 
                act=cached_lookup(col_key) # look up the activity
                inventory_results['amount'].append(lca.inventory[row, col]) 
                inventory_results['elementary_flow'].append(ef.get('name'))
                inventory_results['EF_compartment'].append(ef.get('categories'))
                inventory_results['EF_unit'].append(ef.get('unit'))
                inventory_results['activity_name'].append(act.get('name'))
                inventory_results['activity_location'].append(act.get('location'))
                inventory_results['reference_product'].append(act.get('reference product'))
                
    return inventory_results
```

As before, we run the function to regionalize the inventory, but this time we run it on the full list of activity and automatically save each inventory in the folder. Note that this might take hours, if not days, depending on the size of your list of products.

```python
import re

def make_safe_filename(text, max_length=120):
    # Replace all invalid Windows filename characters
    text = re.sub(r'[<>:"/\\|?*]', '--', text)
    
    # Optional: normalize whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    
    # Avoid trailing dots or spaces (invalid in Windows)
    text = text.rstrip('. ')
    
    # Truncate to avoid path length issues
    return text[:max_length]
```

Next, we can define a function to match each item in an inventory to the appropriate characterization factors.

```python
# New function including LC-Impact


def mapper_land(inv_df, GLAM_land_pivot, mapping_table_GLAM, mapping_table_IW, mapping_table_LC, geo_mapping, iwplus_land, LC_land): 
 
    # Map the ecoinvent inventory data to GLAM flow names
    inv_df_mapped=pd.merge(inv_df, mapping_table_GLAM[['GLAM flow','Ecoinvent','Type','Type2']], how='left', left_on='elementary_flow', right_on='Ecoinvent')
    inv_df_mapped=inv_df_mapped.drop(columns=['Ecoinvent'])
    #inv_df_mapped.to_excel(r"your path\inv_df_mapped.xlsx")

    # Map to geography codes
    inv_df_geo=pd.merge(inv_df_mapped, geo_mapping[['Shortname','GLAM_geo_code','IW+_geo_code','LC_geo_code']], how='left', left_on='activity_location', right_on='Shortname')
    inv_df_geo=inv_df_geo.drop(columns=['Shortname'])
    #inv_df_geo.to_excel(r"your path\inv_df_geo.xlsx")

    # Map to GLAM land use characterization factors
    inv_df_CFs_GLAM=pd.merge(inv_df_geo, GLAM_land_pivot, how='left', left_on=['GLAM flow','Type2', 'GLAM_geo_code'], right_index=True)
    #inv_df_CFs_GLAM.to_excel(r"your path\inv_df_CFs_GLAM.xlsx")

    # Map to IW+ flows
    inv_df_CFs_GLAM_IWmap=pd.merge(inv_df_CFs_GLAM, mapping_table_IW[['IW+','ecoinvent']], how='left', left_on='elementary_flow', right_on='ecoinvent')
    inv_df_CFs_GLAM_IWmap=inv_df_CFs_GLAM_IWmap.drop(columns=['ecoinvent'])
    inv_df_CFs_GLAM_IWmap['IW+_Flow']=inv_df_CFs_GLAM_IWmap['IW+']+', '+inv_df_CFs_GLAM_IWmap['IW+_geo_code'] # Build the full name of the IW+ flow by combining the IW+ flow name and the geography code
    #inv_df_CFs_GLAM_IWmap.to_excel(r"your path\inv_df_CFs_GLAM_IWmap.xlsx")
    
    # Map to IW+ characterization factors
    inv_df_CFs_iw=pd.merge(inv_df_CFs_GLAM_IWmap, iwplus_land[['Elem flow name', 'CF value']], how='left', left_on='IW+_Flow', right_on='Elem flow name')
    inv_df_CFs_iw=inv_df_CFs_iw.drop(columns=['Elem flow name'])
    inv_df_CFs_iw=inv_df_CFs_iw.rename(columns={'CF value':'CF_IW+'})
    #inv_df_CFs_iw.to_excel(r"your path\inv_df_CFs_iw.xlsx")

    
    # Map the ecoinvent inventory data to LC flow names
    inv_df_LC_flows=pd.merge(inv_df_CFs_iw, mapping_table_LC[['LC flow','Ecoinvent']], how='left', left_on='elementary_flow', right_on='Ecoinvent')
    inv_df_LC_flows=inv_df_LC_flows.drop(columns=['Ecoinvent'])
    #inv_df_LC_flows.to_excel(r"your path\inv_df_LC_flows.xlsx")

    # Map to LC land use characterization factors
    inv_df_CFs_LC=pd.merge(inv_df_LC_flows, LC_land, how='left', on=['LC flow','Type2', 'LC_geo_code'])
    #inv_df_CFs_LC.to_excel(r"your path\inv_df_CFs_LC.xlsx")

    
    
    for i in inv_df_CFs_LC.index: # The name of the GLAM and LC flows does not indicate whether they are transformations "to" or "from" a specific land use type, so we need to check the Type column and invert the CFs if necessary.
        if str(inv_df_CFs_LC.loc[i,'Type'])=='from':
            inv_df_CFs_LC.loc[i,'Average Regional CF']=-inv_df_CFs_LC.loc[i,'Average Regional CF']
            inv_df_CFs_LC.loc[i,'Average Global CF']=-inv_df_CFs_LC.loc[i,'Average Global CF']
            inv_df_CFs_LC.loc[i,'Marginal Regional CF']=-inv_df_CFs_LC.loc[i,'Marginal Regional CF']
            inv_df_CFs_LC.loc[i,'Marginal Global CF']=-inv_df_CFs_LC.loc[i,'Marginal Global CF']
            inv_df_CFs_LC.loc[i,'Average Infinite CF']=-inv_df_CFs_LC.loc[i,'Average Infinite CF']
            inv_df_CFs_LC.loc[i,'Average 100y CF']=-inv_df_CFs_LC.loc[i,'Average 100y CF']
            inv_df_CFs_LC.loc[i,'Marginal Infinite CF']=-inv_df_CFs_LC.loc[i,'Marginal Infinite CF']
            inv_df_CFs_LC.loc[i,'Marginal 100y CF']=-inv_df_CFs_LC.loc[i,'Marginal 100y CF']

    inv_df_impact=inv_df_CFs_LC
    inv_df_impact['Average Regional Impact']=inv_df_impact['amount']*inv_df_impact['Average Regional CF']
    inv_df_impact['Average Global Impact']=inv_df_impact['amount']*inv_df_impact['Average Global CF']
    inv_df_impact['Marginal Regional Impact']=inv_df_impact['amount']*inv_df_impact['Marginal Regional CF']
    inv_df_impact['Marginal Global Impact']=inv_df_impact['amount']*inv_df_impact['Marginal Global CF']
    inv_df_impact['Impact_IW+']=inv_df_impact['amount']*inv_df_impact['CF_IW+']
    inv_df_impact['Average Infinite Impact']=inv_df_impact['amount']*inv_df_impact['Average Infinite CF']
    inv_df_impact['Average 100y Impact']=inv_df_impact['amount']*inv_df_impact['Average 100y CF']
    inv_df_impact['Marginal Infinite Impact']=inv_df_impact['amount']*inv_df_impact['Marginal Infinite CF']
    inv_df_impact['Marginal 100y Impact']=inv_df_impact['amount']*inv_df_impact['Marginal 100y CF']
    
    return inv_df_impact
```

## Changing geographies

The following code will change the location of all activities for n number of products. For forestry products up to 5 or in some cases just 1 product covers a large percentage of flows, whereas for concrete and aggregates 5-10 products may be required to cover a similar coverage.

```python
# This function changes the geography for all activities of the n main reference flows

def geo_remapper_topN(inventory_df, geo_code, top_n):

    # 1. Filter land flows
    land_df = inventory_df[
        inventory_df['EF_compartment'].str.contains('land', na=False)
    ].copy()

    # 2. Aggregate total land flow per product
    product_totals = (
        land_df
        .groupby('reference_product')['amount']
        .sum()
        .sort_values(ascending=False)
    )

    # 3. Select top N products
    top_products = product_totals.head(top_n).index

    # 4. Assign geography
    mask = land_df['reference_product'].isin(top_products)
    land_df.loc[mask, 'activity_location'] = geo_code

    return land_df
```

```python
# use this one to just change all geographies for all reference flows, for instance to change all to global CFs

def geo_remapper_all(inventory_df, geo_code):

    # Filter land flows
    land_df = inventory_df[
        inventory_df['EF_compartment'].str.contains('land', na=False)
    ].copy()

    # Assign the same geography to all rows
    land_df['activity_location'] = geo_code

    return land_df
```

## Changing forestry type

The following code will create new inventories where all occurences of extensive forestry are replaced with intensive forestry, and others where it's the opposite. It will results in two sets, representing extensive forestry and intensive forestry.

The adopted approach changes only forestry activities occuring in Sweden. It changes all occupation flows to intensive/extensive forestry, and the net transformation flows to intensive/extensive, see below for some examples:

Extensive dataset: 10 m2 from extensive, 10 m2 to extensive ---> 10 m2 from extensive, 10 m2 to extensive. Cancelling flows, no results change.

Extensive dataset: 10 m2 from intensive, 10 m2 to intensive ---> 10 m2 from extensive, 10 m2 to extensive. Cancelling flows, no results change.

Extensive dataset: 10 m2 from extensive, 10 m2 to intensive ---> 10 m2 from extensive, 10 m2 to extensive. Lowers the impact of extensive dataset.

Extensive dataset: 10 m2 from intensive, 10 m2 to extensive ---> 10 m2 from intensive, 10 m2 to extensive. Keeps the original (beneficial) impact.

Intensive dataset: 10 m2 from extensive, 10 m2 to extensive ---> 10 m2 from intensive, 10 m2 to intensive. Cancelling flows, no results change.

Intensive dataset: 10 m2 from intensive, 10 m2 to intensive ---> 10 m2 from intensive, 10 m2 to intensive. Cancelling flows, no results change.

Intensive dataset: 10 m2 from extensive, 10 m2 to intensive ---> 10 m2 from extensive, 10 m2 to intensive. Keeps the original (damaging) impact.

Intensive dataset: 10 m2 from intensive, 10 m2 to extensive ---> 10 m2 from intensive, 10 m2 to intensive. Rises the impact of intensive dataset.


```python
# This code creates datasets for intensive and extensive forestry from a regionalized inventory, only for forestry activities in Sweden.

# This defines in- and output paths 

path_forest_inventories = r"your path\Remapped_inv_forestry_all\\"
geo_forestry_files = os.listdir(path_forest_inventories)
geo_inventory_files = os.listdir(path_geo_inventories)


# This creates a list of products to be processed.
for prod_name in geo_inventory_files:
    
    activity, geography, unit = prod_name.replace('.xlsx','').split('__')
    
    output_intensive = f"{activity}__{geography}__{unit}__intensive.xlsx"
    output_extensive = f"{activity}__{geography}__{unit}__extensive.xlsx"

    
    # This skips already remapped products in the output folder
    
    if output_intensive not in geo_forestry_files:
        
        inv_df = pd.read_excel(path_geo_inventories + prod_name)

        intensive_df = inv_df.copy()
        extensive_df = inv_df.copy()

        # --------------------------------------------------
        # 1. Filter relevant rows to consider only forestry in Sweden
        # --------------------------------------------------
        mask = (
            inv_df['elementary_flow'].str.contains('forest, intensive|forest, extensive', na=False) &
            (inv_df['activity_location'] == 'SE')
        )

        df_sub = inv_df[mask].copy()

        # --------------------------------------------------
        # 2. Set all occupation flows to intensive or extensive forestry
        # --------------------------------------------------
        occ_mask = df_sub['elementary_flow'].str.startswith('Occupation, forest', na=False)

        # Intensive version
        intensive_df.loc[df_sub[occ_mask].index, 'elementary_flow'] = (
            intensive_df.loc[df_sub[occ_mask].index, 'elementary_flow']
            .str.replace('forest, extensive', 'forest, intensive', regex=False)
        )

        # Extensive version
        extensive_df.loc[df_sub[occ_mask].index, 'elementary_flow'] = (
            extensive_df.loc[df_sub[occ_mask].index, 'elementary_flow']
            .str.replace('forest, intensive', 'forest, extensive', regex=False)
        )

        # --------------------------------------------------
        # 3. Manage transformation flows
        # --------------------------------------------------
        trans_mask = df_sub['elementary_flow'].str.startswith('Transformation', na=False)
        df_trans = df_sub[trans_mask].copy()

        # Group by amount to identify cancelling from/to flows.
        for amount, group in df_trans.groupby(inv_df['amount'].round(10)):

            idxs = group.index
            flows = group['elementary_flow'].tolist()

            has_to = any('Transformation, to forest' in f for f in flows)
            has_from = any('Transformation, from forest' in f for f in flows)

            # ----------------------------------------------
            # CASE 1: Matching pair (to + from, same amount)
            # ----------------------------------------------
            if has_to and has_from and len(group) == 2:
                
                # Replace BOTH
                intensive_df.loc[idxs, 'elementary_flow'] = (
                    intensive_df.loc[idxs, 'elementary_flow']
                    .str.replace('forest, extensive', 'forest, intensive', regex=False)
                )

                extensive_df.loc[idxs, 'elementary_flow'] = (
                    extensive_df.loc[idxs, 'elementary_flow']
                    .str.replace('forest, intensive', 'forest, extensive', regex=False)
                )

            # ----------------------------------------------
            # CASE 2: Same amount but not exact pair
            # ----------------------------------------------
            elif len(group) > 1:
                
                to_idxs = group[group['elementary_flow'].str.contains('Transformation, to forest')].index

                # Replace ONLY "to" transformation flow
                intensive_df.loc[to_idxs, 'elementary_flow'] = (
                    intensive_df.loc[to_idxs, 'elementary_flow']
                    .str.replace('forest, extensive', 'forest, intensive', regex=False)
                )

                extensive_df.loc[to_idxs, 'elementary_flow'] = (
                    extensive_df.loc[to_idxs, 'elementary_flow']
                    .str.replace('forest, intensive', 'forest, extensive', regex=False)
                )

            # ----------------------------------------------
            # CASE 3: No matching amount, replace intensive/extensive only if it is a "to" transformation flow
            # ----------------------------------------------
            else:
                row = group.iloc[0]
                idx = row.name

                if 'Transformation, to forest' in row['elementary_flow']:
                    
                    intensive_df.loc[idx, 'elementary_flow'] = (
                        intensive_df.loc[idx, 'elementary_flow']
                        .replace('forest, extensive', 'forest, intensive')
                    )

                    extensive_df.loc[idx, 'elementary_flow'] = (
                        extensive_df.loc[idx, 'elementary_flow']
                        .replace('forest, intensive', 'forest, extensive')
                    )

                # if "from" → do nothing

        # --------------------------------------------------
        # 4. EXPORT
        # --------------------------------------------------
        intensive_df.to_excel(os.path.join(path_forest_inventories, output_intensive), index=False)
        extensive_df.to_excel(os.path.join(path_forest_inventories, output_extensive), index=False)

        print(f"Saved: {output_intensive} and {output_extensive}")
```
